# What is Data Analysis?

**Data Analysis** is the process of collecting, cleaning, transforming, and interpreting data to discover useful information, identify patterns, and support decision-making.

The main goal of data analysis is to convert raw data into meaningful insights.

## Steps of Data Analysis

1. Data Collection
2. Data Cleaning
3. Data Transformation
4. Data Exploration
5. Data Visualization
6. Interpretation and Decision Making

## Importance of Data Analysis

- Improves decision making
- Identifies trends and patterns
- Helps in prediction and forecasting
- Supports business growth

# How to Plan a Data Science Project?

A Data Science project should follow a structured approach.

## 1. Business Understanding

Define the problem and project objectives.

**Example:**  
Predict whether a customer will default on a loan.

## 2. Data Collection

Gather data from different sources such as:

- CSV files
- Databases
- APIs
- JSON files

## 3. Data Understanding

Analyze:

- Data types
- Missing values
- Outliers
- Data distribution

## 4. Data Preprocessing

- Handle missing values
- Remove outliers
- Encode categorical variables
- Scale features

## 5. Feature Engineering

Create useful features from existing data.

## 6. Model Building

Train Machine Learning models.

## 7. Evaluation

Measure model performance using metrics.

## 8. Deployment

Deploy the model for real-world use.

# How to Frame a Machine Learning Problem?

Framing a Machine Learning problem means converting a business problem into a machine learning task.

## Step 1: Define Objective

**Example:**  
Predict loan default.

## Step 2: Identify Target Variable

**Target:** `default_flag`

- 0 = No Default
- 1 = Default

## Step 3: Identify Features

### Input Variables

- age
- gender
- annual_income
- loan_amount
- credit_score
- repayment_history
- transaction_count
- spending_ratio

## Step 4: Select ML Type

Since the target variable has:

- 0 or 1

It is a:

**Binary Classification Problem**

## Step 5: Define Evaluation Metrics

- Accuracy
- Precision
- Recall
- F1 Score

# Explain Tensors with NumPy Examples

## Definition

A **Tensor** is a mathematical structure used to represent data in multiple dimensions.

A tensor can be:

| Tensor Type | Example |
|------------|----------|
| Scalar | Single Number |
| Vector | 1D Array |
| Matrix | 2D Array |
| Higher Order Tensor | 3D+ Array |

---

## 0-D Tensor (Scalar)

```python
import numpy as np

scalar = np.array(10)

print(scalar)
print(scalar.ndim)
```

### Output

```python
10
0
```

---

## 1-D Tensor (Vector)

```python
vector = np.array([10, 20, 30, 40])

print(vector)
print(vector.ndim)
```

### Output

```python
[10 20 30 40]
1
```

---

## 2-D Tensor (Matrix)

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix)
print(matrix.ndim)
```

### Output

```python
[[1 2 3]
 [4 5 6]]

2
```

---

## 3-D Tensor

```python
tensor3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(tensor3d.ndim)
```

### Output

```python
3
```
